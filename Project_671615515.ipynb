We only want to know whether student dropout or not. Therefore, drop the Enrolled data which is encoded as 1.

```python
data.drop(data[data['Target'] == 1].index, inplace = True)
data
```

Create a new column or feature called 'Dropout' and add the target data Dropout as 1 and Graduate as 0

```python
data['Dropout'] = data['Target'].apply(lambda x: 1 if x==0 else 0)
data.head()
```

```python
import matplotlib.pyplot as plt
data['Dropout'].value_counts().plot(kind='bar')
plt.title('Distribution of Target Variable')
plt.xlabel('Target Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Graduate', 'Dropout'], rotation=0)
plt.show()
```

I will use those data which are correlate with Target with corr value over 1, for independent variables, X.
And Standardize using StandarScaler().fit_Transform()

```python
X = data[["Previous qualification (grade)",
          "Displaced",
          "Admission grade",
          "Curricular units 1st sem (enrolled)",
          "Curricular units 2nd sem (enrolled)",
          "Scholarship holder",
          "Tuition fees up to date",
          "Curricular units 1st sem (approved)",
          "Curricular units 1st sem (grade)",
          "Curricular units 2nd sem (approved)",
          "Curricular units 2nd sem (grade)"]].values






print(X)
X = StandardScaler().fit_transform(X)
X
```

Splitting the data into train dataset and test dataset with the ratio of 8:2

```python
from sklearn.model_selection import train_test_split

y = data['Dropout'].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify= y, random_state=42)
```

# KNeighbors Classifer

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

Prediction form the generated model

```python
y_pred = model.predict(X_test)
print(y_pred)
```

Evaluate the test results with the accuracy value

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

Evaluate the test results with the confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Evaluate test results with Precision, Recall, F1-Score

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
```
